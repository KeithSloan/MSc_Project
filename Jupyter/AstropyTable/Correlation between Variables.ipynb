```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from astropy.table import Table, vstack, join
subSet = Table.read('../../GAMA_Data/REMassEClassEmeasure.fits')
#print(subSet.colnames)
subSet.keep_columns(['Z_1','logmstar','uminusr','absmag_u','DistanceTo5nn','CountInCyl', \
                                          'AGEDenPar','metal','DistanceToEdge'])
print(subSet.colnames)
pd_data = subSet.to_pandas()
corr = pd_data.corr()
fig, ax = plt.subplots(1, 1, figsize=(10,8)) # make the figure with the size 10 x 8 inches
#fig = plt.figure()
ax = fig.add_subplot(111)
cax = ax.matshow(corr,cmap='coolwarm', vmin=-1, vmax=1)
fig.colorbar(cax)
ticks = np.arange(0,len(pd_data.columns),1)
ax.set_xticks(ticks)
plt.xticks(rotation=90)
ax.set_yticks(ticks)
ax.set_xticklabels(pd_data.columns)
ax.set_yticklabels(pd_data.columns)
plt.show()

```
