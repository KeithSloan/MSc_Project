```python
from astropy.table import Table, join
import numpy as np
import matplotlib.pyplot as plt

visualMorph = Table.read('../../GAMA_Data/VisualMorphologyv03.fits')
stellarMasses = Table.read('../../GAMA_Data/StellarMassesv19.fits')
envClass = Table.read('../../GAMA_DATA/GalaxiesClassifiedv01.fits')
envMeasures = Table.read('../../GAMA_DATA/EnvironmentMeasuresv05.fits')
Galvio = Table.read('../../GAMA_DATA/G3CGalv10.fits')
ForGroup = Table.read('../../GAMA_DATA/G3CFoFGroupv10.fits')
print('Visual morph table size : '+str(len(visualMorph)))
# Elliptical code = 1 - Elliptical
#visualMorph1 = visualMorph[visualMorph['ELLIPTICAL_CODE'] == 1]
DataLocalGroup = join(visualMorph,stellarMasses,keys='CATAID',join_type='inner')
print('Visual Morph + StellarMasses '+str(len(DataLocalGroup)))
DataLocalGroup = join(DataLocalGroup,envMeasures,keys='CATAID',join_type='inner')
print('Joining envMeasures '+str(len(DataLocalGroup)))
DataLocalGroup = join(DataLocalGroup,envClass,keys='CATAID',join_type='inner')
print('Joining Environment Measures '+str(len(DataLocalGroup)))
DataLocalGroup = join(DataLocalGroup,Galvio,keys='CATAID',join_type='inner')
print('Joing GalVio '+str(len(DataLocalGroup)))
DataLocalGroup = join(DataLocalGroup,ForGroup,keys='GroupID',join_type='inner')
print('Joing ForGroup '+str(len(DataLocalGroup)))

DataLocalGroup.write('../../GAMA_Data/Data+LocalGroup.fits', format='fits',overwrite=True)


print(DataLocalGroup.colnames)
```
