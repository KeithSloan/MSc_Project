```python
from astropy.table import Table, join
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats
import seaborn as sns
import pandas as pd
import statistics as st


StellarMasses = Table.read('../../GAMA_DATA/StellarMassesv19.fits')
StallarMasses = StellarMasses[StellarMasses['logmstar'] > 0]
print(f"StellarMasses : {len(StellarMasses)}")
meanMass = st.mean(StellarMasses['logmstar'])
print(meanMass)
print(np.mean(StellarMasses['logmstar']))
print(np.max(StellarMasses['logmstar']))
varMass = np.var(StellarMasses['logmstar'])
print(varMass)
StellarMasses = StellarMasses[StellarMasses['logmstar'] > 10]

envMeasures = Table.read('../../GAMA_DATA/EnvironmentMeasuresv05.fits')
print(f"EnvMeasures : {len(envMeasures)}")
envMeasures = envMeasures[envMeasures['SurfaceDensityFlag'] == 0]
envMeasures = envMeasures[envMeasures['SurfaceDensity'] < 15]
envMeasures = envMeasures[envMeasures['AGEDenParFlag'] == 0]
envMeasures = envMeasures[envMeasures['CountInCylFlag'] == 0]
envMeasures = envMeasures[envMeasures['CountInCyl'] > 0]
print(f"EnvMeasures After Cleaning : {len(envMeasures)}")

GroupGal = Table.read('../../GAMA_Data/G3CGalv10.fits')
print(f"GroupGal: {len(GroupGal)}")
# Remove Isolated Galaxies
GroupGal = GroupGal[GroupGal['GroupID'] != 0]
print(f"GroupGal: {len(GroupGal)}")

visualMorph = Table.read('../../GAMA_Data/VisualMorphologyv03.fits')
print(f"visualMorph : {len(visualMorph)}")
visualMorph = visualMorph[visualMorph['ELLIPTICAL_CODE'] == 1]
print(f"visualMorph Elliptical subset: {len(visualMorph)}")

DataLocalGroup1 = join(GroupGal,visualMorph,keys='CATAID',join_type='inner')
print(f"Join GroupGal & VisualMorph Elliptical {len(DataLocalGroup1)}")

DataLocalGroup2 =  join(DataLocalGroup1,envMeasures,keys='CATAID',join_type='inner')
print(f"Join GroupGal & envMeasures {len(DataLocalGroup1)}")

RErange = join(DataLocalGroup2,StellarMasses,keys='CATAID',join_type='inner')
print(f"Join GroupGal - StellarMasses : {len(RErange)}")

yfield = 'uminusr'
alphaVal = .3


fig = plt.figure(figsize=(10, 6), dpi=300)
#fig.suptitle('Plot - Redness versus Environments Galaxies  for Elliptical Galaxies')
xfields = ['CountInCyl','DistanceTo5nn','SurfaceDensity','AGEDenPar']
y = RErange[yfield]
for i,xfield in enumerate(xfields,1) :
    ax = fig.add_subplot(1, len(xfields), i)
    ax.set_ylabel(yfield)
    ax.set_xlabel(xfield)
    #ax.invert_yaxis()
    x = RErange[xfield]
    ax.scatter(x,y,alpha=alphaVal,label=xfield)
    r, p = stats.spearmanr(x,y)
    print(f"Spearman Correlation x: {xfield} y: {yfield} Correlation : {r} T-Test : {p}")
    r, p = stats.pearsonr(x,y)
    print(f"Pearson Correlation x: {xfield} y: {yfield} Correlation : {r} T-Test : {p}")
#fig.legend(loc="upper right")
#plt.axis([0, 3, 0, 0.8])
fig.tight_layout()
#fig.savefig('../../ChartsPlots/AbsRedScatterEnvironments.png', dpi=300, bbox_inches='tight')                      
plt.show()


fig = plt.figure(figsize=(12, 6), dpi=200)
fig.suptitle('Plot - Redness versus Log(Environments) Galaxies  for Elliptical Galaxies')
xfields = ['CountInCyl','DistanceTo5nn','SurfaceDensity','AGEDenPar']
i = 1
y = RErange[yfield]
for xfield in xfields :
    ax = fig.add_subplot(1, len(xfields), i)
    i += 1           
    ax.set_ylabel(yfield)
    ax.set_xlabel('Log10(' +xfield+')')
    #ax.invert_yaxis()
    x = np.log10(RErange[xfield])
    ax.scatter(x,y,alpha=alphaVal,label=xfield)
    
#fig.legend(loc="upper right")
#plt.axis([0, 3, 0, 0.8])
fig.tight_layout()
#fig.savefig('../../ChartsPlots/AbsRedScatterLogEnvironments.png', dpi=300, bbox_inches='tight')
plt.show()

#RErange1 = REMassEClassEmeasure
fig = plt.figure(figsize=(12, 4), dpi=200)
#fig.suptitle('Density Plots - Redness versus Environments Galaxies  for Elliptical Galaxies')
xfields = ['CountInCyl','DistanceTo5nn','SurfaceDensity','AGEDenPar']
y = RErange[yfield]
ymin = min(y)
print(ymin)
ymax = max(y)
print(ymax)
for i,xfield in enumerate(xfields,1) :
    x = RErange[xfield]
    xmin = min(x)
    xmax = max(x)
    X, Y = np.mgrid[xmin:xmax:100j, ymin:ymax:100j]
    positions = np.vstack([X.ravel(), Y.ravel()])
    ax = fig.add_subplot(1, len(xfields), i)
    ax.set_ylabel(yfield)
    ax.set_xlabel(xfield)
    #ax.invert_yaxis()
    values = np.vstack([x, y])
    kernel = stats.gaussian_kde(values)
    Z = np.reshape(kernel(positions).T, X.shape)
    #ax.imshow(np.rot90(Z), cmap=plt.cm.gist_earth_r,extent=[xmin, xmax, ymin, ymax])
    #ax.imshow(np.rot90(Z), cmap='PuBu',extent=[xmin, xmax, ymin, ymax])
    #ax.imshow(np.rot90(Z), cmap='PuBu')
    ax.contour(X,Y,Z)
    #ax.plot(x, y, 'k.', markersize=2)
    #ax.set_xlim([xmin, xmax])
    #ax.set_ylim([ymin, ymax])
    r, p = stats.spearmanr(x,y)
    print(f"Spearman Correlation x: {xfield} y: {yfield} Correlation : {r} T-Test : {p}")
    r, p = stats.pearsonr(x,y)
    print(f"Pearson Correlation x: {xfield} y: {yfield} Correlation : {r} T-Test : {p}")
plt.show()                            
#fig.legend(loc="upper right")
#plt.axis([0, 3, 0, 0.8])
fig.tight_layout()
#fig.savefig('../../ChartsPlots/AbsRedDensityEnvironments.png', dpi=300, bbox_inches='tight')
plt.show()

fig = plt.figure(figsize=(12, 4), dpi=200)
fig.suptitle('Density Plots - High Mass(logmstar>10) uminusr versus Log10(Environments) Galaxies  for Elliptical Galaxies')
xfields = ['CountInCyl','DistanceTo5nn','SurfaceDensity','AGEDenPar']
y = RErange[yfield]
ymin = min(y)
print(ymin)
ymax = max(y)
print(ymax)
for i,xfield in enumerate(xfields,1) :
    x = np.log10(RErange[xfield]+0.01)
    xmin = min(x)
    xmax = max(x)
    X, Y = np.mgrid[xmin:xmax:100j, ymin:ymax:100j]
    positions = np.vstack([X.ravel(), Y.ravel()])
    ax = fig.add_subplot(1, len(xfields), i)  
    ax.set_ylabel(yfield)
    ax.set_xlabel('Log10('+xfield+')')
    #ax.invert_yaxis()
    values = np.vstack([x, y])
    kernel = stats.gaussian_kde(values)
    Z = np.reshape(kernel(positions).T, X.shape)
    #ax.imshow(np.rot90(Z), cmap=plt.cm.gist_earth_r,extent=[xmin, xmax, ymin, ymax])
    #ax.imshow(np.rot90(Z), cmap='PuBu',extent=[xmin, xmax, ymin, ymax])
    #ax.imshow(np.rot90(Z), cmap='PuBu')
    ax.contour(X,Y,Z)
    #ax.plot(x, y, 'k.', markersize=2)
    #ax.set_xlim([xmin, xmax])
    #ax.set_ylim([ymin, ymax])
    r, p = stats.spearmanr(x,y)
    print(f"Log Spearman Correlation x: {xfield} y: {yfield} Correlation : {r} T-Test : {p}")
    r, p = stats.pearsonr(x,y)
    print(f"Log Pearson Correlation x: {xfield} y: {yfield} Correlation : {r} T-Test : {p}")                            
#fig.legend(loc="upper right")
#plt.axis([0, 3, 0, 0.8])
fig.tight_layout()
fig.savefig('../../ChartsPlots/DensityPlot_HighMass_uminusr_Log10_Environments.png', dpi=300, bbox_inches='tight')
plt.show()

#fig, axes = plt.subplots(2, 3, figsize=(18, 10))
fig, axes = plt.subplots(1, 4, figsize=(12, 8), sharey=True)
#fig = plt.figure(figsize=(12, 8), dpi=200)
#fig.suptitle('Countour Plots - Redness versus Environments Galaxies  for Elliptical Galaxies')
xfields = ['CountInCyl','DistanceTo5nn','SurfaceDensity','AGEDenPar']     
y = RErange[yfield]
df = RErange.to_pandas()
for i,xfield in enumerate(xfields) :
    #sns.kdeplot(df[xfield], df.uminusr, ax = axes[i], cmap='Blues', fill=True)
    sns.kdeplot(df[xfield], df[yfield], ax = axes[i], cmap='Blues', fill=False)
    axes[i].invert_yaxis()
    r, p = stats.spearmanr(x,y)
    print(f"Spearman Correlation x: {xfield} y: {yfield} Correlation : {r} T-Test : {p}")
    r, p = stats.pearsonr(x,y)
    print(f"Pearson Correlation x: {xfield} y: {yfield} Correlation : {r} T-Test : {p}")
fig.tight_layout()
#fig.savefig('../../ChartsPlots/AbsRedDensity2Environments.png', dpi=300, bbox_inches='tight')
plt.show()                            


```
