```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import  gennorm, johnsonsu, norm

fig = plt.figure(figsize=(12, 8), dpi=200)
ax1 = fig.add_subplot(1, 2, 1)

beta = 2     # beta = 2 normal dist
binCount = 100
xn = np.linspace(gennorm.ppf(0.01, beta),gennorm.ppf(0.99, beta), binCount)
#print(xn)
#ax1.plot(xn)
xxn = gennorm.pdf(xn, beta)
#print(xx)
ax1.plot(xxn)

r = 9.336
s = 3.548
xj = np.linspace(johnsonsu.ppf(0.01,r,s),johnsonsu.ppf(0.99,r,s),binCount)
xxj = johnsonsu.pdf(xj,r,s)
ax1.plot(xxj)
y = xxn - xxj
#print(y)
ax1.plot(y)

ax2 = fig.add_subplot(1, 2, 2)
mean1 = 10
std1 = 10
x1 = np.linspace(norm.ppf(0.01,mean1,std1),norm.ppf(0.99,mean1,std1),binCount)
xx1 = norm.pdf(x1,mean1,std1)
ax2.plot(xx1)

mean2 = 15
std2 = 20
x2 = np.linspace(norm.ppf(0.01,mean2,std2),norm.ppf(0.99,mean2,std2),binCount)
xx2 = norm.pdf(x2,mean2,std2)
ax2.plot(xx2)





plt.show()
```
