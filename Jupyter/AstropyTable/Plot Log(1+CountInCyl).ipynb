```python
from astropy.table import Table, join
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

RedEllipticalMasses = Table.read('../../GAMA_Data/REMassEClassEmeasure.fits')
REclean = RedEllipticalMasses[RedEllipticalMasses['CountInCyl'] > -500]
#print(RedEllipticalMasses.colnames)
#print(len(RedEllipticalMasses))

alphaVal = 0.3

#### Plot of Redness versus Log(1+CountInCyl)  for Elliptical Galaxies
fig, ax = plt.subplots(1, 1, figsize=(12,8)) # make the figure with the size 10 x 8 inches
fig.suptitle("1 - Plot of Redness Versus Log(1+CountInCyl) for Elliptical Galaxies")
xvar = 'CountInCyl'
yvar = 'uminusr'
# elliminate not read
ax.set_xlabel('Log10( 1+ '+xvar+')')
ax.set_ylabel(yvar)
CinCval = REclean[xvar] + 1
#CinCval = REclean[xvar]
xLog = np.log10(CinCval)
#xLog = CinCval
yLog = REclean[yvar]
ax.scatter(xLog,yLog,alpha=alphaVal,label=yvar)
fig.legend(loc="upper right")
#plt.axis([1.1*min(xLog), 1.1*max(xLog), 1.1*max(yLog), 0.9*min(yLog)])
plt.show()

#### Plot of logmstar versus Log(1+CountinCyl)  for Elliptical Galaxies
fig, ax = plt.subplots(1, 1, figsize=(12,8)) # make the figure with the size 10 x 8 inches
fig.suptitle("2 - Plot of logmass versus Log(1+CountInCyl) for Elliptical Galaxies")
yvar = 'logmstar'
yLog = REclean[yvar]
ax.set_xlabel('Log10( 1+ '+xvar+')')
ax.set_ylabel(yvar)
ax.scatter(xLog,yLog,alpha=alphaVal,label=yvar)
fig.legend(loc="upper right")
#plt.axis([8.6, 11.75, 0.9*min(y), 1.1*max(y)])
plt.show()

#### Plot of log(metal) versus Log(1+CountInCyl)  for Elliptical Galaxies
fig, ax = plt.subplots(1, 1, figsize=(12,8)) # make the figure with the size 10 x 8 inches
fig.suptitle("2 - Plot of log(metal) versus Log(1+CountInCyl) for Elliptical Galaxies")
yvar = 'metal'
yLog = np.log10(REclean[yvar])
ax.set_xlabel('Log10( 1+ '+xvar+')')
ax.set_ylabel(yvar)
ax.scatter(xLog,yLog,alpha=alphaVal,label='Log('+yvar+')')
fig.legend(loc="upper right")
#plt.axis([8.6, 11.75, 0.9*min(y), 1.1*max(y)])
plt.show()




```
