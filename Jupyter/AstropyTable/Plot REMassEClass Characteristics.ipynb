```python
from astropy.table import Table, join
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

RedEllipticalMasses = Table.read('../../GAMA_Data/REMassEClassEmeasure.fits')
print(RedEllipticalMasses.colnames)
print(len(RedEllipticalMasses))

alphaVal = 0.3


##### Plot of Redness versus Redshift for Elliptical Galaxies
fig, ax = plt.subplots(1, 1, figsize=(12,8)) # make the figure with the size 10 x 8 inches
fig.suptitle("2 - Plot of Redness versus Redshift for Elliptical Galaxies")
ax.set_xlabel('Redshift Z')
ax.set_ylabel('Redness (uminusr)')
x = RedEllipticalMasses['Z_1']
print(max(x))
y = RedEllipticalMasses['uminusr']
ax.scatter(x,y,alpha=alphaVal,label='Redness')
fig.legend(loc="upper right")
plt.axis([0.9*min(x), 1.1*max(x), 0.9*min(y), 1.1*max(y)])
plt.show()

##### Plot of LogMass versus Red Shift Z
fig, ax = plt.subplots(1, 1, figsize=(12,8)) # make the figure with the size 10 x 8 inches
fig.suptitle("3 - Plot of LogMass versus Red Shift Z for Elliptical Galaxies")
ax.set_xlabel('Redshift Z')
ax.set_ylabel('Log Mass')
x = RedEllipticalMasses['Z_1']
y = RedEllipticalMasses['logmstar']
ax.scatter(x,y,alpha=alphaVal,label='logmstar')
fig.legend(loc="upper right")
plt.axis([0.9*min(x), 1.1*max(x), 0.9*min(y), 1.1*max(y)])
plt.show()

#### Plot of Count in Cyl versus Red Shift for Elliptical Galaxies
fig, ax = plt.subplots(1, 1, figsize=(12,8)) # make the figure with the size 10 x 8 inches
fig.suptitle("4 - Plot of Count in Cyl versus Red Shift for Elliptical Galaxies")
xvar = 'Z_1'
yvar = 'CountInCyl'
ax.set_xlabel(xvar)
ax.set_ylabel(yvar)
x = RedEllipticalMasses[xvar]
y = RedEllipticalMasses[yvar]
ax.scatter(x,y,alpha=alphaVal,label=yvar)
plt.axis([0.9*min(x), 1.1*max(x), 0, 40])
plt.show()

#### Plot of Distance to 5nn versus Red Shift for Elliptical Galaxies
fig, ax = plt.subplots(1, 1, figsize=(12,8)) # make the figure with the size 10 x 8 inches
fig.suptitle("5 - Plot of Distance to 5nn versus Red Shift for Elliptical Galaxies")
xvar = 'Z_1'
yvar = 'DistanceTo5nn'
ax.set_xlabel(xvar)
ax.set_ylabel(yvar)
x = RedEllipticalMasses[xvar]
y = RedEllipticalMasses[yvar]
ax.scatter(x,y,alpha=alphaVal,label=yvar)
plt.axis([0.9*min(x), 1.1*max(x), 0.9*min(y), 1.1*max(y)])
plt.show()

#### Plot of Metal versus Red Shift for Elliptical Galaxies
fig, ax = plt.subplots(1, 1, figsize=(12,8)) # make the figure with the size 10 x 8 inches
fig.suptitle("6 - Plot of Metal versus Red Shift for Elliptical Galaxies")
xvar = 'Z_1'
yvar = 'metal'
ax.set_xlabel(xvar)
ax.set_ylabel(yvar)
x = RedEllipticalMasses[xvar]
y = RedEllipticalMasses[yvar]
ax.scatter(x,y,alpha=alphaVal,label=yvar)
plt.axis([0.9*min(x), 1.1*max(x), 0.9*min(y), 1.1*max(y)])
plt.show()

##### Plot of Metal versus Redness (uminusr) for Elliptical Galaxies
fig, ax = plt.subplots(1, 1, figsize=(12,8)) # make the figure with the size 10 x 8 inches
fig.suptitle("7 - Plot of Metal versus Redness (uminusr) for Elliptical Galaxies")
ax.set_xlabel('uminusr')
ax.set_ylabel('Metal')
x = RedEllipticalMasses['uminusr']
y = RedEllipticalMasses['metal']
ax.scatter(x,y,alpha=alphaVal,label='metal')
fig.legend(loc="upper right")
plt.axis([1.8, 2.8, 0.9*min(y), 1.1*max(y)])
plt.show()

##### Plot of Metal versus Log Mass for Elliptical Galaxies
fig, ax = plt.subplots(1, 1, figsize=(12,8)) # make the figure with the size 10 x 8 inches
fig.suptitle("8 - Plot of Metal versus Log Mass for Elliptical Galaxies")
ax.set_xlabel('Logmstar')
ax.set_ylabel('Metal')
x = RedEllipticalMasses['logmstar']
y = RedEllipticalMasses['metal']
ax.scatter(x,y,alpha=alphaVal,label='metal')
fig.legend(loc="upper right")
plt.axis([8.6, 11.75, 0.9*min(y), 1.1*max(y)])
plt.show()

#### Plot of Metal versus Log Mass for Elliptical Galaxies
fig, ax = plt.subplots(1, 1, figsize=(12,8)) # make the figure with the size 10 x 8 inches
fig.suptitle("9 - Plot of Redness versus Log Mass for Elliptical Galaxies")
ax.set_xlabel('Logmstar')
ax.set_ylabel('uminusr')
x = RedEllipticalMasses['logmstar']
y = RedEllipticalMasses['uminusr']
ax.scatter(x,y,alpha=alphaVal,label='uminusr')
fig.legend(loc="upper right")
plt.axis([0.9*min(x), 1.1*max(x), 0.9*min(y), 1.1*max(y)])
plt.show()

#### Plot of Redness versus Local Distance for Elliptical Galaxies
fig, ax = plt.subplots(1, 1, figsize=(12,8)) # make the figure with the size 10 x 8 inches
fig.suptitle("10 - Plot of Redness versus Local Distance for Elliptical Galaxies")
ax.set_xlabel('DistanceTo5nn')
ax.set_ylabel('uminusr')
x = RedEllipticalMasses['DistanceTo5nn']
y = RedEllipticalMasses['uminusr']
ax.scatter(x,y,alpha=alphaVal,label='uminusr')
plt.axis([0.9*min(x), 1.1*max(x), 0.9*min(y), 1.1*max(y)])
plt.show()

#### Plot of Log Mass versus Local Distance for Elliptical Galaxies
fig, ax = plt.subplots(1, 1, figsize=(12,8)) # make the figure with the size 10 x 8 inches
fig.suptitle("11 - Plot of Log Mass versus Local Distance for Elliptical Galaxies")
xvar = 'DistanceTo5nn'
yvar = 'logmstar'
ax.set_xlabel(xvar)
ax.set_ylabel(yvar)
x = RedEllipticalMasses[xvar]
y = RedEllipticalMasses[yvar]
ax.scatter(x,y,alpha=alphaVal,label=yvar)
plt.axis([0.9*min(x), 1.1*max(x), 0.9*min(y), 1.1*max(y)])
plt.show()


#### Plot of Metal versus Local Distance for Elliptical Galaxies
fig, ax = plt.subplots(1, 1, figsize=(12,8)) # make the figure with the size 10 x 8 inches
fig.suptitle("12 - Plot of Metal versus Local Distance for Elliptical Galaxies")
ax.set_xlabel('DistanceTo5nn')
ax.set_ylabel('metal')
x = RedEllipticalMasses['DistanceTo5nn']
y = RedEllipticalMasses['metal']
ax.scatter(x,y,alpha=alphaVal,label='metal')
plt.axis([0.9*min(x), 1.1*max(x), 0.9*min(y), 1.1*max(y)])
plt.show()

#### Plot of Surface Density versus Local Distance for Elliptical Galaxies
fig, ax = plt.subplots(1, 1, figsize=(12,8)) # make the figure with the size 10 x 8 inches
fig.suptitle("13 - Plot of Surface Density versus Local Distance for Elliptical Galaxies")
ax.set_xlabel('DistanceTo5nn')
ax.set_ylabel('SurfaceDensity')
x = RedEllipticalMasses['DistanceTo5nn']
y = RedEllipticalMasses['SurfaceDensity']
ax.scatter(x,y,alpha=alphaVal,label='SurfaceDensity')
plt.axis([0, 1.1*max(x), 0, 700])
plt.show()

#### Plot of Redness Density versus CountInCyl for Elliptical Galaxies
fig, ax = plt.subplots(1, 1, figsize=(12,8)) # make the figure with the size 10 x 8 inches
fig.suptitle("14 - Plot of Redness versus Count In Cylinder for Elliptical Galaxies")
xvar = 'CountInCyl'
yvar = 'uminusr'
ax.set_xlabel(xvar)
ax.set_ylabel(yvar)
x = RedEllipticalMasses[xvar]
y = RedEllipticalMasses[yvar]
ax.scatter(x,y,alpha=alphaVal,label=yvar)
plt.axis([0, 40, 0.9*min(y), 1.1*max(y)])
plt.show()

#### Plot of Log Mass Density versus CountInCyl for Elliptical Galaxies
fig, ax = plt.subplots(1, 1, figsize=(12,8)) # make the figure with the size 10 x 8 inches
fig.suptitle("15 - Plot of Log Mass versus Count In Cylinder for Elliptical Galaxies")
xvar = 'CountInCyl'
yvar = 'logmstar'
ax.set_xlabel(xvar)
ax.set_ylabel(yvar)
x = RedEllipticalMasses[xvar]
y = RedEllipticalMasses[yvar]
ax.scatter(x,y,alpha=alphaVal,label=yvar)
plt.axis([0, 40, 0.9*min(y), 1.1*max(y)])
plt.show()

#### Plot of Metal versus CountinCyl for Elliptical Galaxies
fig, ax = plt.subplots(1, 1, figsize=(12,8)) # make the figure with the size 10 x 8 inches
fig.suptitle("16 - Plot of Metal versus Count In Cylinder for Elliptical Galaxies")
xvar = 'CountInCyl'
yvar = 'metal'
ax.set_xlabel(xvar)
ax.set_ylabel(yvar)
x = RedEllipticalMasses[xvar]
y = RedEllipticalMasses[yvar]
ax.scatter(x,y,alpha=alphaVal,label=yvar)
plt.axis([0, 40, 0.9*min(y), 1.1*max(y)])
plt.show()

#### Plot of SurfaceDensity versus CountInCylfor Elliptical Galaxies
fig, ax = plt.subplots(1, 1, figsize=(12,8)) # make the figure with the size 10 x 8 inches
fig.suptitle("17 - Plot of SurfaceDensity versus Count In Cylinder for Elliptical Galaxies")
xvar = 'CountInCyl'
yvar = 'SurfaceDensity'
ax.set_xlabel(xvar)
ax.set_ylabel(yvar)
x = RedEllipticalMasses[xvar]
y = RedEllipticalMasses[yvar]
ax.scatter(x,y,alpha=alphaVal,label=yvar)
plt.axis([0, 40, 0, 1000])
plt.show()
```
