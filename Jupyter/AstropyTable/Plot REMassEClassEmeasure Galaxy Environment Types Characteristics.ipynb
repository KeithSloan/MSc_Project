```python
from astropy.table import Table, join
import numpy as np
import matplotlib.pyplot as plt

REMassEClassEmeasure = Table.read('../../GAMA_Data/REMassEClassEmeasure.fits')
#print(REMassEClassEmeasure.colnames)
REGalaxyVoids = REMassEClassEmeasure[REMassEClassEmeasure['GeoS4']== 0]
print('Red Elliptical Galaxy Voids : '+str(len(REGalaxyVoids)))
REGalaxySheet = REMassEClassEmeasure[REMassEClassEmeasure['GeoS4']== 1]
print('Red Elliptical Galaxy Sheets : '+str(len(REGalaxySheet)))
REGalaxyFilament = REMassEClassEmeasure[REMassEClassEmeasure['GeoS4']== 2]
print('Red Elliptical Galaxy Filament : '+str(len(REGalaxyFilament)))
REGalaxyKnot = REMassEClassEmeasure[REMassEClassEmeasure['GeoS4']== 3]
print('Red Elliptical Galaxy Knot : '+str(len(REGalaxyKnot)))

###################### Redness
xfield = 'Z_1'
yfield = 'uminusr'
alphaVal = .3
x1 = REGalaxyVoids[xfield]
y1 = REGalaxyVoids[yfield]
x2 = REGalaxySheet[xfield]
y2 = REGalaxySheet[yfield]
x3 = REGalaxyFilament[xfield]
y3 = REGalaxyFilament[yfield]
x4 = REGalaxyKnot[xfield]
y4 = REGalaxyKnot[yfield]
fig, ax = plt.subplots(1, 1, figsize=(12,6)) # make the figure with the size 10 x 6 inches
fig.suptitle('Plot 1 - Redness (uminusr) versus Red Shift Z for Elliptical Galaxies')
ax.set_ylabel('Redness usr')
ax.set_xlabel('Redshift Z')
ax.scatter(x1,y1,alpha=alphaVal,label='Voids')
ax.scatter(x2,y2,alpha=alphaVal,label='Sheet')
ax.scatter(x3,y3,alpha=alphaVal,label='Filament')
ax.scatter(x4,y4,alpha=alphaVal,label='knot')
fig.legend(loc="upper right")
#plt.axis([0, 3, 0, 0.8])
plt.show()

###################### Metallicity
xfield = 'Z_1'
yfield = 'metal'
alphaVal = .3
x1 = REGalaxyVoids[xfield]
y1 = REGalaxyVoids[yfield]
x2 = REGalaxySheet[xfield]
y2 = REGalaxySheet[yfield]
x3 = REGalaxyFilament[xfield]
y3 = REGalaxyFilament[yfield]
x4 = REGalaxyKnot[xfield]
y4 = REGalaxyKnot[yfield]
fig, ax = plt.subplots(1, 1, figsize=(12,6)) # make the figure with the size 10 x 6 inches
fig.suptitle('Plot 2 - Metal versus Red Shift Z for Elliptical Galaxies')
ax.set_ylabel('Metal')
ax.set_xlabel('Redshift Z')
ax.scatter(x1,y1,alpha=alphaVal,label='Voids')
ax.scatter(x2,y2,alpha=alphaVal,label='Sheet')
ax.scatter(x3,y3,alpha=alphaVal,label='Filament')
ax.scatter(x4,y4,alpha=alphaVal,label='knot')
fig.legend(loc="upper right")
#plt.axis([0, 3, 0, 0.8])
plt.show()

###################### log Mass
xfield = 'Z_1'
yfield = 'logmstar'
alphaVal = .3
x1 = REGalaxyVoids[xfield]
y1 = REGalaxyVoids[yfield]
x2 = REGalaxySheet[xfield]
y2 = REGalaxySheet[yfield]
x3 = REGalaxyFilament[xfield]
y3 = REGalaxyFilament[yfield]
x4 = REGalaxyKnot[xfield]
y4 = REGalaxyKnot[yfield]
fig, ax = plt.subplots(1, 1, figsize=(12,6)) # make the figure with the size 10 x 6 inches
fig.suptitle('Plot 3 -Log Mass versus Red Shift Z for Elliptical Galaxies')
ax.set_ylabel('logmstar')
ax.set_xlabel('Redshift Z')
ax.scatter(x1,y1,alpha=alphaVal,label='Voids')
ax.scatter(x2,y2,alpha=alphaVal,label='Sheet')
ax.scatter(x3,y3,alpha=alphaVal,label='Filament')
ax.scatter(x4,y4,alpha=alphaVal,label='knot')
fig.legend(loc="upper right")
#plt.axis([0, 3, 0, 0.8])
plt.show()



######### Group Findings
Groups = Table.read('../../GAMA_Data/GroupsClassified.fits')
G3Groups = Table.read('../../GAMA_Data/G3CGalv10.fits')

REGroupKnot=join(REGalaxyKnot,G3Groups,keys='CATAID',join_type='inner')
REGroupFilament = join(REGalaxyFilament,G3Groups,keys='CATAID',join_type='inner')
REGroupSheet = join(REGalaxySheet,G3Groups,keys='CATAID',join_type='inner')
REGroupVoids = join(REGalaxyVoids,G3Groups,keys='CATAID',join_type='inner')

xfield = 'SepIterCen'
yfield = 'uminusr'
alphaVal = .3
x1 = REGroupVoids[xfield]
y1 = REGroupVoids[yfield]
x2 = REGroupSheet[xfield]
y2 = REGroupSheet[yfield]
x3 = REGroupFilament[xfield]
y3 = REGroupFilament[yfield]
x4 = REGroupKnot[xfield]
y4 = REGroupKnot[yfield]
fig, ax = plt.subplots(1, 1, figsize=(12,6)) # make the figure with the size 10 x 6 inches
fig.suptitle('Plot 4 - Redness versus Angular Separation to Group Centre')
ax.set_ylabel(yfield)
ax.set_xlabel(xfield)
ax.scatter(x1,y1,alpha=alphaVal,label='Voids')
ax.scatter(x2,y2,alpha=alphaVal,label='Sheet')
ax.scatter(x3,y3,alpha=alphaVal,label='Filament')
ax.scatter(x4,y4,alpha=alphaVal,label='knot')
fig.legend(loc="upper right")
plt.axis([0, 1500, 1.8, 3])
plt.show()

xfield = 'RankIterCen'
yfield = 'uminusr'
alphaVal = .3
x1 = REGroupVoids[xfield]
y1 = REGroupVoids[yfield]
x2 = REGroupSheet[xfield]
y2 = REGroupSheet[yfield]
x3 = REGroupFilament[xfield]
y3 = REGroupFilament[yfield]
x4 = REGroupKnot[xfield]
y4 = REGroupKnot[yfield]
fig, ax = plt.subplots(1, 1, figsize=(12,6)) # make the figure with the size 10 x 6 inches
fig.suptitle('Plot 5 - Redness versus Rank to Centre Centre')
ax.set_ylabel(yfield)
ax.set_xlabel(xfield)
ax.scatter(x1,y1,alpha=alphaVal,label='Voids')
ax.scatter(x2,y2,alpha=alphaVal,label='Sheet')
ax.scatter(x3,y3,alpha=alphaVal,label='Filament')
ax.scatter(x4,y4,alpha=alphaVal,label='knot')
fig.legend(loc="upper right")
plt.axis([0, 200, 1.8, 3])
plt.show()

xfield = 'SepIterCen'
yfield = 'logmstar'
alphaVal = .3
x1 = REGroupVoids[xfield]
y1 = REGroupVoids[yfield]
x2 = REGroupSheet[xfield]
y2 = REGroupSheet[yfield]
x3 = REGroupFilament[xfield]
y3 = REGroupFilament[yfield]
x4 = REGroupKnot[xfield]
y4 = REGroupKnot[yfield]
fig, ax = plt.subplots(1, 1, figsize=(12,6)) # make the figure with the size 10 x 6 inches
fig.suptitle('Plot 6 - Redness versus Angular Separation to Group Centre')
ax.set_ylabel(yfield)
ax.set_xlabel(xfield)
ax.scatter(x1,y1,alpha=alphaVal,label='Voids')
ax.scatter(x2,y2,alpha=alphaVal,label='Sheet')
ax.scatter(x3,y3,alpha=alphaVal,label='Filament')
ax.scatter(x4,y4,alpha=alphaVal,label='knot')
fig.legend(loc="upper right")
plt.axis([0, 1500, 8.5, 11.5])
plt.show()

xfield = 'RankIterCen'
yfield = 'logmstar'
alphaVal = .3
x1 = REGroupVoids[xfield]
y1 = REGroupVoids[yfield]
x2 = REGroupSheet[xfield]
y2 = REGroupSheet[yfield]
x3 = REGroupFilament[xfield]
y3 = REGroupFilament[yfield]
x4 = REGroupKnot[xfield]
y4 = REGroupKnot[yfield]
fig, ax = plt.subplots(1, 1, figsize=(12,6)) # make the figure with the size 10 x 6 inches
fig.suptitle('Plot 7 - Log Mass versus Rank to Centre Centre')
ax.set_ylabel(yfield)
ax.set_xlabel(xfield)
ax.scatter(x1,y1,alpha=alphaVal,label='Voids')
ax.scatter(x2,y2,alpha=alphaVal,label='Sheet')
ax.scatter(x3,y3,alpha=alphaVal,label='Filament')
ax.scatter(x4,y4,alpha=alphaVal,label='knot')
fig.legend(loc="upper right")
plt.axis([0, 150, 8.5, 11.5])
plt.show()

xfield = 'SepIterCen'
yfield = 'metal'
alphaVal = .3
x1 = REGroupVoids[xfield]
y1 = REGroupVoids[yfield]
x2 = REGroupSheet[xfield]
y2 = REGroupSheet[yfield]
x3 = REGroupFilament[xfield]
y3 = REGroupFilament[yfield]
x4 = REGroupKnot[xfield]
y4 = REGroupKnot[yfield]
fig, ax = plt.subplots(1, 1, figsize=(12,6)) # make the figure with the size 10 x 6 inches
fig.suptitle('Plot 8 - Redness versus Angular Separation to Group Centre')
ax.set_ylabel(yfield)
ax.set_xlabel(xfield)
ax.scatter(x1,y1,alpha=alphaVal,label='Voids')
ax.scatter(x2,y2,alpha=alphaVal,label='Sheet')
ax.scatter(x3,y3,alpha=alphaVal,label='Filament')
ax.scatter(x4,y4,alpha=alphaVal,label='knot')
fig.legend(loc="upper right")
#plt.axis([0, 1500, 1.8, 3])
plt.axis([0, 1500, 0, 0.05])
plt.show()

xfield = 'RankIterCen'
yfield = 'metal'
alphaVal = .3
x1 = REGroupVoids[xfield]
y1 = REGroupVoids[yfield]
x2 = REGroupSheet[xfield]
y2 = REGroupSheet[yfield]
x3 = REGroupFilament[xfield]
y3 = REGroupFilament[yfield]
x4 = REGroupKnot[xfield]
y4 = REGroupKnot[yfield]
fig, ax = plt.subplots(1, 1, figsize=(12,6)) # make the figure with the size 10 x 6 inches
fig.suptitle('Plot 9 - Log Mass versus Rank to Centre Centre')
ax.set_ylabel(yfield)
ax.set_xlabel(xfield)
ax.scatter(x1,y1,alpha=alphaVal,label='Voids')
ax.scatter(x2,y2,alpha=alphaVal,label='Sheet')
ax.scatter(x3,y3,alpha=alphaVal,label='Filament')
ax.scatter(x4,y4,alpha=alphaVal,label='knot')
fig.legend(loc="upper right")
plt.axis([0, 200, 0, 0.05])
plt.show()
```
