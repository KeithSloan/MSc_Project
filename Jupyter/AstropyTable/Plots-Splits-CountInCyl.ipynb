```python
from astropy.table import Table, join
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

RedEllipticalMasses = Table.read('../../GAMA_Data/REMassEClassEmeasure.fits')
xvar = 'CountInCyl'
REclean = RedEllipticalMasses[RedEllipticalMasses[xvar] > -500]
#print(RedEllipticalMasses.colnames)
#print(len(RedEllipticalMasses))

alphaVal = 0.3
RElow1 = REclean[REclean[xvar]>0]
RElow = RElow1[RElow1[xvar]<3]

REmid1 = REclean[REclean[xvar]>= 3]
REmid = REmid1[REmid1[xvar]<=7]

REhigh1 = REclean[REclean[xvar]>7]
REhigh  = REhigh1[REhigh1[xvar]<=12]

REVhigh = REclean[REclean[xvar]>12]

#### Plot of Redness versus Log(1+CountInCyl)  for Elliptical Galaxies
fig, ax = plt.subplots(1, 1, figsize=(12,8)) # make the figure with the size 10 x 8 inches
fig.suptitle("1 - Plot of Redness Versus log Mass for sets based on CountInCyl")
xvar = 'logmstar'
yvar = 'uminusr'
# elliminate not read
ax.set_xlabel(xvar)
ax.set_ylabel(yvar)
xlog1 = RElow[xvar]
ylog1 = RElow[yvar]
ylog2 = REmid[yvar]
xlog2 = REmid[xvar]
ylog3 = REhigh[yvar]
xlog3 = REhigh[xvar]
ylog4 = REVhigh[yvar]
xlog4 = REVhigh[xvar]
#ax1.scatter(x[:4], y[:4], s=10, c='b', marker="s", label='first')
#ax1.scatter(x[40:],y[40:], s=10, c='r', marker="o", label='second')
ax.scatter(xlog1,ylog1,alpha=alphaVal,label='Low')
ax.scatter(xlog2,ylog2,alpha=alphaVal,label='Mid')
ax.scatter(xlog3,ylog3,alpha=alphaVal,label='High')
ax.scatter(xlog4,ylog4,alpha=alphaVal,label='V High')
fig.legend(loc="upper right")
#plt.axis([1.1*min(xLog), 1.1*max(xLog), 1.1*max(yLog), 0.9*min(yLog)])
plt.show()

fig, ax = plt.subplots(1, 1, figsize=(12,8)) # make the figure with the size 10 x 8 inches
fig.suptitle("2 - Stacked Histogram Counts of Red Galaxies for sets based on DistanceTo5nn")
ax.set_xlabel('uminusr')
ax.set_ylabel('Counts of Galaxies')
#Stack the data
#plt.figure()
ax.hist([ylog1,ylog2,ylog3,ylog4], bins=50, stacked=True, density=True)
ax.legend({ "Low", "Mid",  "High",  "V High"}) 
#fig.legend(loc="upper right")
plt.show()





```
