```python
from astropy.table import Table, join
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

RedEllipticalMasses = Table.read('../../GAMA_Data/REMassEClassEmeasure.fits')
REclean = RedEllipticalMasses[RedEllipticalMasses['CountInCyl'] > -500]
#print(RedEllipticalMasses.colnames)
#print(len(RedEllipticalMasses))


xvar = 'DistanceTo5nn'
alphaVal = 0.3
RElow1 = REclean[REclean[xvar]>0]
RElow = RElow1[RElow1[xvar]<1.7]

REmid1 = REclean[REclean[xvar]>= 1.7]
REmid = REmid1[REmid1[xvar]<=3]

REhigh = REclean[REclean[xvar]>3]

#### Plot of Redness versus Log(1+CountInCyl)  for Elliptical Galaxies
fig, ax = plt.subplots(1, 1, figsize=(12,8)) # make the figure with the size 10 x 8 inches
fig.suptitle("1 - Plot of Redness Versus log Mass for sets based on DistanceTo5nn")
xvar = 'logmstar'
yvar = 'uminusr'
# elliminate not read
ax.set_xlabel(xvar)
ax.set_ylabel(yvar)
xlog1 = RElow[xvar]
ylog1 = RElow[yvar]
ylog2 = REmid[yvar]
xlog2 = REmid[xvar]
ylog3 = REhigh[yvar]
xlog3 = REhigh[xvar]
ax.scatter(xlog1,ylog1,alpha=alphaVal,label='Low')
ax.scatter(xlog2,ylog2,alpha=alphaVal,label='Mid')
ax.scatter(xlog3,ylog3,alpha=alphaVal,label='High')
fig.legend(loc="upper right")
#plt.axis([1.1*min(xLog), 1.1*max(xLog), 1.1*max(yLog), 0.9*min(yLog)])
plt.show()

#counts1, bins1 = np.histogram(RElow[yvar],bins=100)
#counts2, bins2 = np.histogram(RElow[yvar],bins=100)
#counts3, bins3 = np.histogram(RElow[yvar],bins=100)
#counts = [counts1, counts2, counts3] 
#centers = [centers for _ in counts] 
fig, ax = plt.subplots(1, 1, figsize=(12,8)) # make the figure with the size 10 x 8 inches
fig.suptitle("2 - Stacked Histogram Counts of Red Galaxies for sets based on DistanceTo5nn")
ax.set_xlabel('uminusr')
ax.set_ylabel('Counts of Galaxies')
#Stack the data
#plt.figure()
ax.hist([ylog1,ylog2,ylog3], bins=50, stacked=True, density=True)
ax.legend({"Low", "Mid", "High"}) 
fig.legend(loc="upper right")
plt.show()






```
